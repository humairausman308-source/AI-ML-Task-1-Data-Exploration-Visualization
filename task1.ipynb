# Task 1: Exploring and Visualizing the Iris Dataset

## Objective
Load, inspect, and visualize the Iris dataset to understand feature distributions and relationships between species.

This notebook includes:
- Dataset loading and inspection
- Exploratory Data Analysis (EDA)
- Scatter plots, histograms, and boxplots
- Summary insights


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')  # Load dataset from seaborn
iris.head()
```

```python
# Shape and column names
print("Shape of dataset:", iris.shape)
print("Columns:", iris.columns)

# Dataset info
iris.info()

# Descriptive statistics
iris.describe()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set1')
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
iris.hist(figsize=(10,8), bins=15)
plt.suptitle("Feature Distributions")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=iris)
plt.title("Box Plot of Iris Features")
plt.show()
```

## Summary / Insights

- The dataset has 150 samples and 5 columns (4 features + species).
- Scatter plots show clear separation of species by petal and sepal dimensions.
- Histograms reveal the distribution of each feature.
- Box plots help identify potential outliers in the dataset.
- This task demonstrates basic data exploration and visualization skills.
